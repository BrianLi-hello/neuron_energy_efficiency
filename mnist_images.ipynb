```python
import matplotlib.pyplot as plt
import torchvision.datasets as datasets
from torchvision.transforms import ToTensor
```

```python
# Load the MNIST dataset
mnist_dataset = datasets.MNIST(root="./dataset", train=False, download=True, transform=ToTensor())

# Create a figure and axes
fig, ax = plt.subplots(2, 5, figsize=(12, 5))

# Iterate through rows
for row in range(2):
    # Iterate through columns
    for col in range(5):
        # Calculate the index of the current digit
        index = row * 5 + col

        # Get an image and its corresponding label
        image, label = mnist_dataset[index]

        # Plot the image on the corresponding axis
        ax[row, col].imshow(image.squeeze(), cmap="gray")
        ax[row, col].axis("off")
        # ax[row, col].set_title(str(label.item()))

# Adjust spacing between subplots
fig.subplots_adjust(hspace=0.4)

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
